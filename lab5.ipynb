1. Implement the functions in equations (1), (2), (3) and (4).

```python
import math
# 1 - function
def func(x):
    return (6 * x - 2) ** 2 * math.sin(12 * x - 4)

# 2 - function
def g(x):
    return math.sin(10 * x)

# 3 - function
def death(x, inifite):
    return func(x) if g(x) <= 0 else infinite

def s(x, r, b):
    return r * max(0, g(x)) ** b

# 4 - function
def static(x, r, b):
    return func(x) + s(x, r, b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 1000)
# function responses
y1 = [func(i) for i in x]
plt.plot(x, y1)
```

```python
y2 = [g(i) for i in x]
plt.plot(x, y2)
```

```python
infinite = 20
y3 = [death(i, infinite) for i in x]
plt.plot(x, y3)
```

```python
r = 20; b = 1
y4 = [static(i, r, b) for i in x]
plt.plot(x, y4)
```
